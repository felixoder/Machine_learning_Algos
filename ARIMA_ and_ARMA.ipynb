```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_series = [12, 14, 15, 18, 19, 21, 20, 22, 24, 26]
```

```python
# Fit ARMA model (ARMA(2,2))
model = ARIMA(time_series, order=(2, 0, 2)) # ARMA with d=0
model_fit = model.fit()
```

```python
print("ARMA Model Summary:")
print(model_fit.summary())
```

```python
# Predictions
predictions = model_fit.predict(start=0, end=len(time_series) + 4)
print("Predicted Values (ARMA):", predictions)
```

```python
# Plot original and predicted values
plt.figure(figsize=(10, 6))
plt.plot(range(len(time_series)), time_series, label="Original Data",
marker='o')
plt.plot(range(len(predictions)), predictions, label="Predictions",
linestyle='--', marker='x')
plt.title("ARMA Model: Predictions vs Original Data")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()
```

```python
# Import libraries
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Example non-stationary time series data
time_series = [1, 2, 4, 8, 16, 32, 64, 128, 256, 512]
```

```python
# Fit ARIMA model (ARIMA(2, 1, 2))
model = ARIMA(time_series, order=(2, 1, 2)) # AR=2, I=1, MA=2
model_fit = model.fit()
```

```python
# Summary of the model
print("ARIMA Model Summary:")
print(model_fit.summary())
```

```python
# Predictions
predictions = model_fit.predict(start=1, end=len(time_series) + 4)
print("Predicted Values (ARIMA):", predictions)
```

```python
# Plot original and predicted values
plt.figure(figsize=(10, 6))
plt.plot(range(len(time_series)), time_series, label="Original Data",
marker='o')
plt.plot(range(len(predictions)), predictions, label="Predictions",
linestyle='--', marker='x')
plt.title("ARIMA Model: Predictions vs Original Data")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()
```
